```python
import json
```

```python
# 验证当前生成的数据问题
data = json.load(open('./doc_fact.json', 'r'))
```

```python
# 类型1为“没有找到”：“该案件中没有与”、“抱歉”、“没有提到”
# 类型2为“重复输出”：有多个“起因、经过、结果分别为：”，多个“起因：”
# 类型2为“不符合格式的输出”：只要不是“起因、经过、结果分别为：”开头或者“该案件没有与”开头的都是有问题的
type1 = []
type2 = []
type3 = []
for did in data.keys():
    for crime in data[did].keys():
        # 类型1
        if any([i in data[did][crime] for i in ['该案件没有与', '抱歉', '没有提到']]):
            type1.append(f'{did}_{crime}')
        # 类型2
        if data[did][crime].count('起因、经过、结果分别为：') > 1 or data[did][crime].count('起因：') > 1 or '起因：无\n经过：无\n结果：无' in data[did][crime]:
            type2.append(f'{did}_{crime}')
        
        if not (data[did][crime].split('：', 1)[0].endswith('起因、经过、结果分别为') or data[did][crime].startswith('该案件中没有与')):
            type3.append(f'{did}_{crime}')
```
